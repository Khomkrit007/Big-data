* (10 points) จากข้อมูลที่ให้มา มีค่าว่างสูญหายมากน้อยแค่ไหน? และมีค่าผิดปกติ (outliers) และหรือข้อมูลรบกวน (noisy data) หรือไม่ในแต่ละตัวแปร?

```python

# เติมค่าที่หายไปใน 'age' ด้วยอายุเฉลี่ย
titanic['age'].fillna(titanic['age'].mean(), inplace=True)

# เติมค่าที่หายไปใน 'embarked' ด้วยค่าที่พบบ่อยที่สุด
titanic['embarked'].fillna(titanic['embarked'].mode()[0], inplace=True)

# เนื่องจาก 'cabin' มีค่าที่หายไปจำนวนมาก ควรลบ column ออก
# ตรวจสอบให้แน่ใจว่าชื่อ column ถูกต้อง
titanic.drop('cabin', axis=1, inplace=True)

# สำหรับ 'boat' และ 'body' ค่าที่หายไปบ่งชี้ว่าผู้โดยสารไม่ได้รอดชีวิต
# เติมค่าที่หายไปใน 'boat' ด้วย 'None' เพื่อระบุว่าไม่มีเรือ
titanic['boat'].fillna('None', inplace=True)

# เติมค่าที่หายไปใน 'fare ด้วย 'None' เพื่อระบุว่าไม่มีข้อมูลการจ่ายค่าตั๋ว
titanic['fare'].fillna('None', inplace=True)

# เติมค่าที่หายไปใน 'body' ด้วย 0 เพื่อระบุว่าไม่พบศพ
# แก้ไขการพิมพ์ผิด: เปลี่ยน 'titanasdc' เป็น 'titanic' และ 'b' เป็น 'body'
titanic['body'].fillna(0, inplace=True)

# เติมค่าที่หายไปใน 'home.dest' ด้วย 'Unknown'
titanic['home.dest'].fillna('Unknown', inplace=True)

# ตรวจสอบว่ามีค่าที่หายไปเหลืออยู่หรือไม่
print(titanic.isnull().sum())
```

* (10 points) จากข้อมูลที่ให้มา เตรียมข้อมูลให้พร้อมโดยการจัดการกับค่าว่างในข้อมูล ให้อธิบายวิธีที่ท่านเลือกใช้ พร้อมยกเหตุผลประกอบ

```python
#ใช้ one-hot encoding ในการแปลงตัวแปร Embarked
titanic = pd.get_dummies(titanic, columns=['embarked'], prefix=['embarked'])
```

```python
# แปลงตัวแปร Sex โดยให้เพศชายเป็น 0 และเพศหญิงเป็น 1

# Create a mapping dictionary for sex
sex_mapping = {'male': 0, 'female': 1}

# Apply the mapping to the 'sex' column
titanic['sex'] = titanic['sex'].map(sex_mapping)
```

```python
#สร้างตัวแปร FamilySize ซึ่งเป็นผลรวมของตัวแปร SibSp (number of siblings/spouses aboard) และ Parch (number of parents/children aboard)

# Create a new column 'FamilySize' by adding 'sibsp' and 'parch' columns.
titanic['FamilySize'] = titanic['sibsp'] + titanic['parch']
```

```python
#สร้างตัวแปร IsAlone ซึ่งมีค่าเป็น 1 ถ้า FamilySize เป็น 0, และมีค่าเป็นศูนย์เมื่อเป็นอื่น

# Create a new column 'IsAlone'
titanic['IsAlone'] = 0

# Set 'IsAlone' to 1 where FamilySize is 0
titanic.loc[titanic['FamilySize'] == 0, 'IsAlone'] = 1
```

* แปลงข้อมูลเชิงคุณภาพให้เป็นตัวเลข \
    i. (10 points) ใช้ one-hot encoding ในการแปลงตัวแปร Embarked \
    ii. (10 points) แปลงตัวแปร Sex โดยให้เพศชายเป็น 0 และเพศหญิงเป็น 1 \
    iii. (10 points) สร้างตัวแปร FamilySize ซึ่งเป็นผลรวมของตัวแปร SibSp (number of siblings/spouses aboard) และ Parch (number of parents/children aboard) \
    iv. (10 points) สร้างตัวแปร IsAlone ซึ่งมีค่าเป็น 1 ถ้า FamilySize เป็น 0, และมีค่าเป็นศูนย์เมื่อเป็นอื่น

```python
#ตัวแปร Fare และ Age ใช้ Min-Max Scaling

from sklearn.preprocessing import MinMaxScaler

# Create a MinMaxScaler object
scaler = MinMaxScaler()

# Handle missing values before scaling (replace with the median for example)
titanic['age'].fillna(titanic['age'].median(), inplace=True)

# Convert 'fare' column to numeric type, coercing errors to NaN
titanic['fare'] = pd.to_numeric(titanic['fare'], errors='coerce')

# Handle missing values in 'fare' after converting to numeric
titanic['fare'].fillna(titanic['fare'].median(), inplace=True)

# Fit the scaler to the Fare and Age columns and transform the data
titanic[['fare', 'age']] = scaler.fit_transform(titanic[['fare', 'age']])
```

```python
# สร้างตัวแปรใหม่ Fare_Per_Person ด้วยการนำตัวแปร Fare มาหารด้วย FamilySize + 1

# Calculate Fare per person
titanic['Fare_Per_Person'] = titanic['fare'] / (titanic['FamilySize'] + 1)
```

```python
#สร้างตัวแปร AgeGroup ที่ระบุกลุ่มอายุของตัวอย่าง โดยแบ่งเป็นกลุ่มดังนี้ [0, 12), [12, 18), [18, 50) และ [50, 100)

# Define a function to categorize age into groups
def age_group(age):
  if age < 0.12:
    return "[0, 12)"
  elif age < 0.18:
    return "[12, 18)"
  elif age < 0.5:
    return "[18, 50)"
  else:
    return "[50, 100)"

# Apply the function to create the AgeGroup column
titanic['AgeGroup'] = titanic['age'].apply(age_group)
```

1. ชั้นโดยสาร (Pclass)
แผนภูมิแรกแสดงจำนวนผู้โดยสารในแต่ละชั้นโดยสาร:
ชั้น 3 มีจำนวนผู้โดยสารมากที่สุด ตามด้วยชั้น 1 และชั้น 2
แผนภูมินี้แสดงให้เห็นถึงการจัดแบ่งผู้โดยสารเป็น 3 ระดับชั้น ซึ่งชั้น 3 มีจำนวนผู้โดยสารหนาแน่นมากที่สุด
แผนภูมิที่สองแสดงจำนวนผู้รอดชีวิตและเสียชีวิตในแต่ละชั้นโดยสาร:
ผู้โดยสารในชั้นที่ 1 มีอัตราการรอดชีวิตสูงที่สุดเมื่อเทียบกับชั้นอื่น ๆ
ชั้นที่ 3 มีจำนวนผู้เสียชีวิตมากที่สุด โดยสัดส่วนของผู้เสียชีวิตนั้นสูงมากเมื่อเทียบกับจำนวนผู้รอดชีวิต
ชั้น 2 มีจำนวนผู้โดยสารน้อยกว่า แต่สัดส่วนผู้รอดชีวิตและผู้เสียชีวิตเกือบเท่ากัน
2. เพศ (Sex)
แผนภูมิที่สามแสดงจำนวนผู้โดยสารตามเพศ:มีผู้โดยสารชายมากกว่าผู้โดยสารหญิง
แสดงให้เห็นถึงความไม่เท่าเทียมกันของจำนวนผู้โดยสารชายและหญิงบนเรือไททานิค
แผนภูมิที่สี่แสดงการรอดชีวิตของผู้โดยสารตามเพศ:

ผู้โดยสารหญิงมีอัตราการรอดชีวิตที่สูงกว่าผู้โดยสารชายอย่างชัดเจน
สัดส่วนผู้หญิงที่รอดชีวิตมีมากกว่าสัดส่วนของผู้หญิงที่เสียชีวิต ในขณะที่ผู้โดยสารชายส่วนใหญ่เสียชีวิต
สะท้อนถึงการช่วยเหลือผู้หญิงก่อนในช่วงเหตุการณ์วิกฤต (นโยบาย "Women and children first")
3. อายุและค่าโดยสาร (Age & Fare)
แผนภูมิที่ห้าเป็นแผนภูมิกระจาย (scatter plot) แสดงความสัมพันธ์ระหว่างอายุและค่าโดยสาร (Fare) ของผู้โดยสาร:
ค่าโดยสารที่สูงกว่ามีแนวโน้มสัมพันธ์กับการรอดชีวิตมากกว่า โดยเฉพาะผู้ที่จ่ายค่าโดยสารสูง (ส่วนใหญ่เป็นผู้โดยสารชั้น 1)
ผู้โดยสารที่อายุไม่มาก (เด็กและคนหนุ่มสาว) มีโอกาสรอดชีวิตสูงกว่าผู้โดยสารที่มีอายุมากกว่า
กลุ่มผู้โดยสารที่จ่ายค่าโดยสารสูง (อายุกลางๆ) รอดชีวิตมากกว่า ส่วนผู้ที่จ่ายค่าโดยสารต่ำ (ส่วนใหญ่อยู่ในชั้น 3) มีแนวโน้มเสียชีวิตมากกว่า
4. กลุ่มอายุ (Age Group)
แผนภูมิสุดท้ายแสดงการรอดชีวิตและเสียชีวิตตามกลุ่มอายุ (AgeGroup):
กลุ่มอายุที่มีตั้งแต่ 0-12 ปี มีโอกาสรอดชีวิตมากที่สุด แสดงให้เห็นถึงความสำคัญในการช่วยเหลือเด็ก
กลุ่มอายุที่มีตั้งแต่ 18-50 ปี มีอัตราการเสียชีวิตสูงกว่ากลุ่มอื่น ๆ ซึ่งอาจสะท้อนให้เห็นถึงกลุ่มผู้โดยสารชายที่มีจำนวนมากในช่วงอายุนี้
กลุ่มอายุ 50 ปีขึ้นไป (สูงอายุ) มีอัตราการเสียชีวิตที่สูงเช่นกัน

การวิเคราะห์เชิงลึก:
ชั้นโดยสารและการรอดชีวิต: ผู้โดยสารชั้นที่ 1 มีโอกาสรอดชีวิตสูงที่สุด ซึ่งอาจเป็นเพราะพวกเขาอยู่ในชั้นที่ใกล้กับดาดฟ้าเรือและได้รับการช่วยเหลือก่อน
เพศและการรอดชีวิต: ผู้หญิงมีอัตราการรอดชีวิตสูงกว่าผู้ชายอย่างมีนัยสำคัญ สะท้อนถึงนโยบายการช่วยเหลือผู้หญิงก่อน
อายุและค่าโดยสาร: ผู้โดยสารที่อายุน้อยและจ่ายค่าโดยสารสูงมีโอกาสรอดชีวิตมากกว่า ซึ่งอาจบ่งชี้ถึงการให้ความสำคัญกับการช่วยเหลือเด็กและผู้โดยสารในชั้นที่ 1
กลุ่มอายุ: กลุ่มเด็กและผู้สูงอายุมีอัตราการรอดชีวิตที่ต่างกันอย่างชัดเจน เด็กมีโอกาสรอดชีวิตสูง แต่ผู้สูงอายุมีอัตราการเสียชีวิตสูง
